```python
from ultralytics import YOLO
model = YOLO("yolov3-tinyu.pt")
```

```python
results = model.train(
    data = "../data/dart_positions/dart_positions.yaml",
    epochs=25,
    imgsz=160,
    batch=8,
    half=True,
    amp=True,
)
```

```python
from pathlib import Path
images = []
for i in range(1, 1000):
    img_train = f"../data/dart_positions/images/train/img_{i:04d}.jpg"
    # img_val = f"../data/dart_positions/images/val/img_{i:04d}.jpg"
    if Path(img_train).exists():
        images.append(img_train)

results = model(images)
for res in results:
    if res.boxes:
        print(res.boxes)
        res.show()
```
